```python
import numpy as np
from math import sqrt, pi, exp
import matplotlib.pyplot as plt
```

# 空間フーリエ変換

```python
N = 4096            # サンプル数
s = N/256           # 標準偏差

y = []
for i in range(N):
  x = i - N/2
  v = exp(-x**2/(2.0*s**2))/(sqrt(2*pi)*s)
  y.append(v)
```

```python
plt.plot(y)
plt.xlim([N/2-100,N/2+100])
plt.show()
```

```python
fk = np.fft.fft(y)
plt.plot(fk)
```

```python
freq = np.fft.fftfreq(N)
plt.plot(freq)
```

```python
# plt.xlim([-0.05,0.05])
# plt.scatter(freq,fk,s=1)
plt.xlim([-0.05,0.05])
plt.scatter(freq,np.abs(fk),s=1)
```

```python
theory = [exp(-(2*pi*k)**2*s**2/2.0) for k in freq]    # 理論曲線
fig, ax = plt.subplots()
ax.scatter(freq, np.abs(fk), s = 1, c = "red")  # DFTの結果
ax.scatter(freq,theory, s = 1, c="blue")        # 理論曲線
plt.xlim([-0.1,0.1])
plt.show()
```

# 時間フーリエ変換

```python
N = 4096            # サンプル数
h = 0.1          # 時間刻み
gamma = 0.2
v0 = 2.0

v = v0
f = []
for _ in range(N):
  v = v -gamma * v * h
  f.append(v)
```

```python
plt.plot(f)
plt.show()
```

```python
freq = np.fft.fftfreq(N, d=h)
plt.plot(freq)
```

```python
fw = np.fft.fft(f)
plt.scatter(freq, np.abs(fw),s=1)
```
